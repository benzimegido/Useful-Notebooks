## Trying Neural Networks

Neural Networks (NN) are often giving better results than traditional statistical models. Let's build a simple NN and try to apply it the Boston dataset. 

```python
Image(url= "https://cdn-images-1.medium.com/max/1600/1*uz3wd5YeVYlU2JR8rE9VDA.png")
```

These hyper parameters are here to be played with and to find better ones

```python
#Define training hyperprameters.
batch_size = 50
num_epochs = 200
learning_rate = 0.01
size_hidden= 100

#Calculate some other hyperparameters based on data.  
batch_no = len(X_train) // batch_size  #batches
cols=X_train.shape[1] #Number of columns in input matrix
n_output=1
```

## Neural Network (NN) Definition

We will build the simplest NN with a single hidden linear layer.

The NN need an **init** function (where we define what are the layers in it) and a **forward** function that describes the flow through the network. As part of the forward function we define which **activation** functions to use (Relu, in this case).

```python
import torch
import torch.nn.functional as F
#Create the model
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
# Assume that we are on a CUDA machine, then this should print a CUDA device:
print("Executing the model on :",device)
class Net(torch.nn.Module):
    def __init__(self, n_feature, size_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(cols, size_hidden)   # hidden layer
        self.predict = torch.nn.Linear(size_hidden, n_output)   # output layer

    def forward(self, x):
        x = F.relu(self.hidden(x))      # activation function for hidden layer
        x = self.predict(x)             # linear output
        return x
net = Net(cols, size_hidden, n_output)
```

```python
Image(url= "https://cdn-images-1.medium.com/max/1600/1*-a-flCLHLCGM0-7TOcNJnQ.png")
```

## Optimizer Definition

Next we need to define which **optimizer** to use (Adam, as default) and what is the **Loss function**.

```python
#Adam is a specific flavor of gradient decent which is typically better
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)
#optimizer = torch.optim.SGD(net.parameters(), lr=0.2)
criterion = torch.nn.MSELoss(reduction='sum')  # this is for regression mean squared loss
#criterion = torch.nn.L1Loss(reduction='sum')  # this is for regression mean absolute loss
```

## NN Training

The stadard flow of training the netword is based on:
* Loading the **mini batch** (lines 11-12)
* **Forward** flow (line 17)
* **Loss** calculation (lines 18-19)
* **Backward** step (line 20)

```python
from sklearn.utils import shuffle
from torch.autograd import Variable
running_loss = 0.0
for epoch in range(num_epochs):
    #Shuffle just mixes up the dataset between epocs
    X_train, Y_train = shuffle(X_train, Y_train)
    # Mini batch learning
    for i in range(batch_no):
        start = i * batch_size
        end = start + batch_size
        inputs = Variable(torch.FloatTensor(X_train[start:end].values))
        labels = Variable(torch.FloatTensor(Y_train[start:end].ravel()))
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, torch.unsqueeze(labels,dim=1))
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        
    print('Epoch {}'.format(epoch+1), "loss: ",running_loss)
    running_loss = 0.0
```

## Error Estimation

```python
X = Variable(torch.FloatTensor(X_test.values)) 
result = net(X)
pred=result.data[:,0].numpy()
print(len(pred),len(Y_test))
```

```python
plt.scatter(Y_test, pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$")
```
